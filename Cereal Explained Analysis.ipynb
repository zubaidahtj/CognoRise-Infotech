The carbohydrates, sugars, and potassium variables have negative values which doesn't make sense in the context of the cereals content. To solve this problem, I will replace the negative values with the median of said variables.

```python
neg_carb = df[df['carbo']<0]
neg_sug = df[df['sugars']<0]
neg_potass = df[df['potass']<0]

neg_carb
```

```python
neg_sug
```

```python
neg_potass
```

```python
median_carb = df['carbo'].median()
median_sug = df['sugars'].median()
median_potass = df['potass'].median()

df['carbo'] = df['carbo'].apply(lambda x: median_carb if x < 0 else x)
df['sugars'] = df['sugars'].apply(lambda x: median_sug if x < 0 else x)
df['potass'] = df['potass']. apply(lambda x: median_potass if x < 0 else x)
```

### Identifying any potential outliers

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

##defining the outliers
outliers = df[((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
outliers
```

```python
##capping outliers at 5th and 95th percentiles
for column in df.columns:
    if df[column].dtype != 'object': 
        lower_bound = df[column].quantile(0.05)
        upper_bound = df[column].quantile(0.95)
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
```

```python
outliers_new = df[((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
outliers_new
```
